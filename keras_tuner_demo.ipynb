```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32,activation='relu', input_dim=8))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=32, epochs=100, validation_data=(X_test, y_test))
```

```python
import kerastuner as kt
```

```python
def build_model(hp):
  model = Sequential()
  model.add(Dense(32,activation='relu', input_dim=8))
  model.add(Dense(1,activation='sigmoid'))

  optimizer = hp.Choice('optimizer', values=['adam','sgd','rmsprob', 'adadelta'])
  model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
  return model

```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5
)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
import shutil
# Remove the old Keras Tuner project directory to ensure a fresh start.
# This step is crucial if you've changed the build_model architecture.
shutil.rmtree('mydir', ignore_errors=True)
print("Removed old 'mydir' directory.")
```

```python
# Re-initialize the Keras Tuner with the build_model function.
# It's important that this build_model reflects the architecture you intend to search.
import keras_tuner as kt

tuner = kt.RandomSearch(build_model,
                        objective = 'val_accuracy',
                        max_trials=3,
                        directory= 'mydir', # This directory is now clean after shutil.rmtree
                        project_name='num_layers')
```

```python
# Re-run the search for the best hyperparameters.
# This will generate new models and save their weights and configurations.
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
model = tuner.get_best_models(num_models=1)[0]
```

```python
# Now, you can retrieve the best model. Since the tuner started fresh,
# the loaded model architecture will match the saved weights.
model = tuner.get_best_models(num_models=1)[0]
model.summary()
```

```python
def build_model(hp):
  model = Sequential()

  counter = 0

  unit = hp.Int('units', min_value=8, max_value=128)

  for i in range(hp.Int('num_layers', min_value=1, max_value=10)):
    if counter == 0:
      model.add(Dense(
          hp.Int('units' + str(i), min_value=8, max_value=128),
          activation=hp.Choice('activation' + str(i), values=['relu', 'tanh', 'sigmoid']),
          input_dim = 8))

    else:
      model.add(Dense(
          hp.Int('units' + str(i), min_value=8, max_value=128),
          activation=hp.Choice('activation' + str(i), values=['relu', 'tanh', 'sigmoid'])))

    counter += 1


  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer = hp.Choice('optimizer', values = ['rmsprop', 'adam', 'sgd', 'nadam', 'adadelta']), loss='binary_crossentropy', metrics=['accuracy'])
  return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective = 'val_accuracy',
                        max_trials=3,
                        directory= 'mydir',
                        project_name='num_layers')
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```
