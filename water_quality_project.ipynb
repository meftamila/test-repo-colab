```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Potability', axis=1),
                                                    df['Potability'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Chloramines'], X_train['Conductivity'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Chloramines'], X_train_scaled['Conductivity'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Solids'], ax=ax1)
sns.kdeplot(X_train['Conductivity'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Solids'], ax=ax2)
sns.kdeplot(X_train_scaled['Conductivity'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Hardness Distribution Before Scaling')
sns.kdeplot(X_train['Hardness'], ax=ax1)

# after scaling
ax2.set_title('Hardness Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Hardness'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Organic_carbon Distribution Before Scaling')
sns.kdeplot(X_train['Organic_carbon'], ax=ax1)

# after scaling
ax2.set_title('Organic_carbon Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Organic_carbon'], ax=ax2)
plt.show()
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train_scaled, y_train)

y_pred = logreg.predict(X_test_scaled)
```

```python
y_pred
```

```python
# import the metrics class
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')

```

```python
from sklearn.metrics import classification_report
target_names = ['without Potability', 'with Potability']
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier( n_estimators=300,
    max_depth=10,
    min_samples_split=4,
    class_weight='balanced',
    random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
y_pred = rf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from xgboost import XGBClassifier

ratio = (y_train == 0).sum() / (y_train == 1).sum()

xgb_model = XGBClassifier(
    scale_pos_weight=ratio,
    learning_rate=0.05,
    n_estimators=400,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train_scaled, y_train)
```
